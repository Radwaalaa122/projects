```python
Merge_data = pd.merge(rating, anime_copy, on='anime_id')
Merge_data
```

```python
Merge_data.describe()
```

```python
Merge_data.rename(columns={'rating_x': 'user_rating'}, inplace=True)
Merge_data.rename(columns={'rating_y': 'rating'}, inplace=True)
```

```python
Merge_data['episodes'] = pd.to_numeric(Merge_data['episodes'], errors='coerce')
```

```python
Merge_data['episodes']=Merge_data['episodes'].astype('Int64')
```

```python
Merge_data.info()
```

```python
Q1_episodes = Merge_data['episodes'].quantile(0.25)
Q3_episodes = Merge_data['episodes'].quantile(0.75)
IQR_episodes = Q3_episodes - Q1_episodes
IQR_episodes
```

```python
lower_episodes = Q1_episodes - 1.5 * IQR_episodes
lower_episodes
```

```python
upper_episodes = Q3_episodes + 1.5 * IQR_episodes
upper_episodes
```

```python
outliers_episodes = Merge_data[(Merge_data['episodes'] < (Q1_episodes - 1.5 * IQR_episodes)) | (Merge_data['episodes'] > (Q3_episodes + 1.5 * IQR_episodes))]
outliers_episodes
```

```python
outliers_episodes = outliers_episodes.dropna()
```

```python
plt.boxplot(outliers_episodes['episodes'])
```

```python
no_outliers = Merge_data[(Merge_data['episodes'] >= lower_episodes) & (Merge_data['episodes'] <= upper_episodes)]
```

```python
plt.boxplot(no_outliers['episodes'])
```
